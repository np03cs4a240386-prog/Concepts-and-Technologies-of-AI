```python
import numpy as np

# Define matrices A and B
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
identity_check

```

```python
AB = np.dot(A, B)
BA = np.dot(B, A)

AB, BA, AB == BA  # Shows AB and BA are not equal
```

```python
AB_T = np.dot(A, B).T
B_T_A_T = np.dot(B.T, A.T)

AB_T, B_T_A_T, np.allclose(AB_T, B_T_A_T)  # True if they are equal
```

```python
# Coefficient matrix
coeff = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3, 1, -1]])

# Constant vector
const = np.array([-1, -3, 9])
solution = np.linalg.solve(coeff, const)
solution
```

```python
coeff_inv = np.linalg.inv(coeff)
solution_inv = np.dot(coeff_inv, const)
solution_inv
```
